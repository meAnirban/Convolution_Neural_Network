```python
import cv2
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
%matplotlib inline
import matplotlib.pyplot as plt

#Reading image as gray scale
CATEGORIES = ["Dog", "Cat"]  # will use this to convert prediction num to string value

def prepare(filepath):
    IMG_SIZE = 50  
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)  # read in the image, convert to grayscale
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))  # resize image to match model's expected sizing
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)  

model = tf.keras.models.load_model("resources/64x3-CNN.model2")
```

```python
img1 = image.load_img('cat_test.jpg')
plt.imshow(img1);
```

```python
img2 = image.load_img('dog_test.png')
plt.imshow(img2);
```

```python
prediction = model.predict([prepare('cat_test.jpg')])
print(CATEGORIES[int(prediction[0][0])])
```

```python
prediction = model.predict([prepare('dog_test.png')])
print(CATEGORIES[int(prediction[0][0])])
```
